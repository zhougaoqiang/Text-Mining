```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "data/Training_Essay_Data.csv"
df = pd.read_csv(file_path)

print("Dataset Overview:")
print(df.info())
```

#### Analysis: datase

```python
print("First 5 rows of the dataset:")
print(df.head(5))

print("Last 5 rows of the dataset:")
print(df.tail(5))
```

```python
label_column = "generated"
text_column = "text"
print("\nSample AI-generated texts:")
print(df[df[label_column] == 1 ].sample(2)[text_column].tolist())

print("\nSample Human-written texts:")
print(df[df[label_column] == 0 ].sample(2)[text_column].tolist())
```

```python
print(df[label_column].value_counts(normalize=True))
#check sample distribution
plt.figure(figsize=(6,4))
sns.countplot(x=df[label_column])
plt.title("Class Distribution")
plt.xticks([0, 1], ['Human', 'Generated'])
plt.xlabel("Category")
plt.ylabel("Sample Count")
plt.show()
```

```python
if text_column in df.columns:
    df["text_length"] = df[text_column].astype(str).apply(len)

    plt.figure(figsize=(8,5))
    sns.histplot(df["text_length"], bins=50, kde=True)
    plt.title("Text Length Distribution")
    plt.xlabel("Text Length")
    plt.ylabel("Sample Count")
    plt.show()

    print("\nText Length Statistics:")
    print(df["text_length"].describe())
```

```python
text_column = "text"
if text_column in df.columns:
    df["word_count"] = df[text_column].astype(str).apply(lambda x: len(x.split()))

    plt.figure(figsize=(8,5))
    sns.histplot(df["word_count"], kde=True)
    plt.title("Word Count Distribution")
    plt.xlabel("Word Count")
    plt.ylabel("Sample Count")
    plt.show()

    print("\nWord Count Statistics:")
    print(df["word_count"].describe())
```
